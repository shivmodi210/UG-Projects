```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv("final_data.csv")

bulk_modulus = data.iloc[:, 12]
D_bulk = data.iloc[:, 13]

X1 = data.iloc[:, 0:12]
X2 = data.iloc[:, 14:106]

X = np.concatenate((X1, X2),axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, bulk_modulus, random_state = 1312, test_size = 0.2, shuffle=True)

regr1 = linear_model.LinearRegression()
regr1.fit(X_train, Y_train)
print(regr1.score(X_train, Y_train))
print(regr1.score(X_test, Y_test))

plt.plot(Y_test, regr1.predict(X_test), 'r.')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, D_bulk, random_state = 1312, test_size = 0.2, shuffle=True)

regr2 = linear_model.LinearRegression()
regr2.fit(X_train, Y_train)
print(regr2.score(X_train, Y_train))
print(regr2.score(X_test, Y_test))

plt.plot(Y_test, regr2.predict(X_test), 'r.')
plt.show()
```
